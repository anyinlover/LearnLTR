```python
from catboost import Pool, CatBoostRanker
```

```python
train_path = "/home/guhangsong/Data/istella-letor/new_train.txt"
test_path = "/home/guhangsong/Data/istella-letor/new_test.txt"
```

```python
train_pool = Pool(train_path, column_description='istella_catboost.cd', delimiter=',')
```

```python
test_pool = Pool(test_path, column_description='istella_catboost.cd', delimiter=',')
```

```python
from catboost.utils import get_gpu_device_count
get_gpu_device_count()
```

```python
from catboost.utils import create_cd
create_cd(label=0, group_id=1, output_path='istella_catboost.cd')
```

```python
model = CatBoostRanker(iterations=2000, custom_metric='NDCG:top=10', task_type='GPU', devices='0')
```

```python
model.fit(train_pool, eval_set=test_pool, plot=True)
```

```python
model = CatBoostRanker(iterations=2000, loss_function='YetiRankPairwise', custom_metric='NDCG:top=10', task_type='GPU', devices='0', used_ram_limit='16gb')
```

```python
model.fit(train_pool, eval_set=test_pool, plot=True)
```

```python
model.eval_metrics(test_pool, 'NDCG:top=10')
```
