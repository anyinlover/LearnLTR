```python
from catboost import Pool, CatBoostRanker
from catboost.utils import create_cd
```

```python
train_path = "/home/guhangsong/Data/yahoo/set1.train.csv"
valid_path = "/home/guhangsong/Data/yahoo/set1.valid.csv"
test_path = "/home/guhangsong/Data/yahoo/set1.test.csv"
```

```python
create_cd(label=0, group_id=1, output_path='yahoo_catboost.cd')
```

```python
train_pool = Pool(train_path, column_description='yahoo_catboost.cd', delimiter=',')
valid_pool = Pool(valid_path, column_description='yahoo_catboost.cd', delimiter=',')
test_pool = Pool(test_path, column_description='yahoo_catboost.cd', delimiter=',')
```

```python
model = CatBoostRanker(iterations=8000, loss_function='YetiRankPairwise', custom_metric='NDCG:top=10', task_type='GPU', devices='0', used_ram_limit='16gb')
```

```python
model.fit(train_pool, eval_set=valid_pool)
```

```python
x = model.eval_metrics(test_pool, 'NDCG:top=10')
```

```python
x['NDCG:top=10;type=Base'][-1]
```

```python
y = model.eval_metrics(valid_pool, 'NDCG:top=10')
y['NDCG:top=10;type=Base'][-1]
```
